```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatText, Dropdown, FileUpload, VBox, Label
from IPython.display import display
```

```python
upload = FileUpload(accept='.jpg,.jpeg,.png', multiple=False)

wall_height_input = FloatText(description='Wall Height (m):', value=5.0)
pit_width_input = FloatText(description='Pit Width (m):', value=6.0)

material_dropdown = Dropdown(
    options=['10mm','20mm','coarse_sand','fine_sand'],
    description='Material:'
)

display(VBox([
    Label("Upload Image"),
    upload,
    Label("Enter Dimensions"),
    wall_height_input,
    pit_width_input,
    material_dropdown
]))
```

```python
def load_image():
    if len(upload.value) == 0:
        print("Upload image first.")
        return None
    
    file_info = upload.value[0]
    content = file_info['content'].tobytes()
    np_img = np.frombuffer(content, np.uint8)
    img = cv2.imdecode(np_img, cv2.IMREAD_COLOR)

    if img is None:
        print("Image decoding failed.")
        return None
    
    return img
```

```python
def segment_sand(img):

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Adjustable HSV range for sand
    lower = np.array([0, 0, 40])
    upper = np.array([180, 100, 220])

    mask = cv2.inRange(hsv, lower, upper)

    kernel = np.ones((7,7), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        print("No sand contour detected.")
        return None

    largest = max(contours, key=cv2.contourArea)

    final_mask = np.zeros(mask.shape, dtype=np.uint8)
    cv2.drawContours(final_mask, [largest], -1, 255, thickness=cv2.FILLED)

    return final_mask
```

```python
def calculate_height_profile(mask):

    h, w = mask.shape
    heights = []

    for x in range(w):
        column = mask[:, x]
        ys = np.where(column == 255)[0]

        if len(ys) > 0:
            top = np.min(ys)
            height = h - top
        else:
            height = 0

        heights.append(height)

    return np.array(heights)
```

```python
def calculate_volume_and_weight(heights, img_shape):

    wall_height = wall_height_input.value
    pit_width = pit_width_input.value
    material = material_dropdown.value

    h, w = img_shape[:2]

    pixel_to_meter_height = wall_height / h
    pixel_to_meter_width = pit_width / w

    heights_m = heights * pixel_to_meter_height

    volume = np.sum(heights_m * pixel_to_meter_width)

    density_map = {
        "10mm": 1600,
        "20mm": 1650,
        "coarse_sand": 1700,
        "fine_sand": 1550
    }

    weight = volume * density_map[material]

    return volume, weight
```

```python
img = load_image()

if img is not None:

    mask = segment_sand(img)

    if mask is not None:

        heights = calculate_height_profile(mask)
        volume, weight = calculate_volume_and_weight(heights, img.shape)

        # Display Results
        plt.figure(figsize=(6,4))
        plt.imshow(mask, cmap='gray')
        plt.title("Final Sand Mask")
        plt.axis("off")
        plt.show()

        plt.figure(figsize=(8,3))
        plt.plot(heights)
        plt.title("Height Profile (Pixel)")
        plt.show()

        print("Estimated Volume (m³):", round(volume,2))
        print("Estimated Weight (tons):", round(weight/1000,2))
```
